# binning特征

```python
data3=pd.read_csv(r'D:\读研\ML\机器学习进阶实战\1.数据特征\数值特征\datasets\fcc_2016_coder_survey_subset.csv',encoding='utf-8')
```

```python
plt.hist(data3['Age'],color='#A9C5D3')
plt.show()
plt.close()
```

```python
data3['Age_cut']=np.array(np.floor(np.array(data3['Age'])/10))
```

# 按中位数进行切分

```python
quantile_list=[0,.25,.75,1]
quatiles=data3['Income'].quantile(quantile_list)
```

```python
fig,ax=plt.subplots()
data3['Income'].hist(bins=30)
for quatile in quatiles:
    qv1=plt.axvline(quatile,color='r')
```

```python
data3['Income_new']=pd.qcut(data3['Income'],q=quantile_list)#按照百分位切分
```

# 对数变换，可以降低偏度值，更贴近正态，一般是对标签做

```python
data3['Income_log']=np.log((1+data3['Income']))
data3[['Income','Income_log']].head()
```

```python
fig,ax1=plt.subplots(2,1)
ax1[0].hist(data3['Income'],bins=30)
ax1[1].hist(data3['Income_log'],bins=30)
plt.show()
```

# 处理日期相关特征，转成pandas格式，可以直接用，Timestamp

```python
time_stamps = ['2015-03-08 10:30:00.360000+00:00', '2017-07-13 15:45:05.755000-07:00',
               '2012-01-20 22:30:00.254000+05:30', '2016-12-25 00:30:00.000000+10:00']
df = pd.DataFrame(time_stamps, columns=['Time'])
ts_objs = np.array([pd.Timestamp(item) for item in np.array(df.Time)])
df['TS_obj'] = ts_objs
ts_objs
```
