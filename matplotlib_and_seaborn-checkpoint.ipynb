```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x=np.linspace(1,500,100)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y,'r--')#第一种rz方式
plt.scatter(x,y)
plt.xlim([0,500])
plt.show()
```

```python
fig=plt.figure()#第二种麻瓜方式
ax=fig.add_subplot(211)
ax.scatter(x,y)
ax1=fig.add_subplot(212)
ax1.plot(x,y)
```

```python
fig,ax=plt.subplots(2,2)#第三种大佬方式
ax[0][0].plot(x,y)
ax[0][0].set_xlim([0,10])
plt.show()
```
